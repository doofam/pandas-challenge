```python
import pandas as pd
```

```python
import os
```

```python
schools_df = pd.read_csv("schools_complete.csv")
students_df = pd.read_csv("students_complete.csv")
```

```python
schools_df
```

```python
students_df
```

```python
df = pd.merge(schools_df, students_df, how="left")
```

```python
df.head(10)
```

```python
## District Summary

# Total Schools 
total_schools = df["school_name"].nunique()
total_schools

# Total Students 
total_students = df["student_name"].count()
total_students

# Total Budget 
total_budget = df["budget"].unique().sum()
total_budget

# Average Math Score 
average_math = df["math_score"].mean().round(5)
average_math

# Average Reading Score
average_reading = df["reading_score"].mean().round(5)
average_reading 

# Percent Passing Math 
passing_math_df = df[df["math_score"] >= 70]
len(passing_math_df)
passing_math = float(len(passing_math_df)/len(df))*100

# Percent Passing Reading
passing_reading_df = df[df["reading_score"] >= 70]
len(passing_reading_df)
passing_reading = float(len(passing_reading_df)/len(df))*100

# Overall Passing Rate = passing_math + passing_reading / 2 
overall_passing = float(passing_math + passing_reading) / 2
overall_passing

# Creating District Summary Table
dict1 = {"Total Schools": [total_schools], 
        "Total Students": [total_students], 
        "Total Budget": ["$" + str(total_budget)],
        "Average Math Score": [average_math],
        "Average Reading Score": [average_reading],
        "% Passing Math": [passing_math],
        "% Passing Reading": [passing_reading],
        "Overall Passing Rate": [overall_passing]}
district_df = pd.DataFrame(dict1)
district_df
```

```python
## School Summary

# School Name, School Type, Total Students, Total School Budget
school_df = schools_df[["school_name", "type", "size", "budget"]].set_index("school_name")

# Per Student Budget
school_df["Per Student Budget"] = school_df["budget"] / school_df["size"]

# Average Math Score
math_score_byschool = df.groupby("school_name")["math_score"].mean().round(5)

# Average Reading Score
reading_score_byschool = df.groupby("school_name")["reading_score"].mean().round(5)

# Percent Passing Math
math_passing_df = df[df["math_score"] >= 70]
students_passing_math_byschool = math_passing_df.groupby("school_name")["math_score"].count()
size_byschool = school_df["size"]
percent_passing_math_byschool = students_passing_math_byschool.divide(size_byschool)*100

# Percent Passing Reading
reading_passing_df = df[df["reading_score"] >= 70]
students_passing_reading_byschool = reading_passing_df.groupby("school_name")["reading_score"].count()
percent_passing_reading_byschool = students_passing_reading_byschool.divide(size_byschool)*100

# Overall Passing Rate
overall_passing_rate_byschool = percent_passing_math_byschool.add(percent_passing_reading_byschool)/2

# Creating School Summary Table
school_df1 = pd.DataFrame({"Average Math Score": math_score_byschool, 
                        "Average Reading Score": reading_score_byschool,
                        "% Passing Math": percent_passing_math_byschool,
                        "% Passing Reading": percent_passing_reading_byschool,
                        "Overall Passing Rate": overall_passing_rate_byschool})

school_summary_df = pd.merge(school_df, school_df1, left_on="school_name", right_on="school_name")

school_summary_df = school_summary_df.rename(columns={"school_name":"School Name", 
                                                      "type":"Type", 
                                                      "size":"Size", 
                                                      "budget":"Budget"})
school_summary_df
```

```python
## Top Performing Schools 
top_performing_df = school_summary_df.sort_values(["Overall Passing Rate"], ascending=False)
top_performing_df[:5]
```

```python
## Bottom Performing Schools
bottom_performing_df = school_summary_df.sort_values(["Overall Passing Rate"])
bottom_performing_df[:5]
```

```python
## Math Scores by Grade

ninth_df = df.loc[df["grade"] == "9th"]
ninth_math_s = ninth_df.groupby("school_name")["math_score"].mean()

tenth_df = df.loc[df["grade"] == "10th"]
tenth_math_s = tenth_df.groupby("school_name")["math_score"].mean()

eleventh_df = df.loc[df["grade"] == "11th"]
eleventh_math_s = eleventh_df.groupby("school_name")["math_score"].mean()

twelveth_df = df.loc[df["grade"] == "12th"]
twelveth_math_s = twelveth_df.groupby("school_name")["math_score"].mean()

math_bygrade_df = pd.DataFrame({"9th": ninth_math_s,
                                "10th": tenth_math_s,
                                "11th": eleventh_math_s,
                                "12th": twelveth_math_s})
math_bygrade_df

# Simpler way (but different format): math_bygrade_df = df.groupby(["school_name", "grade"])["math_score"].mean().to_frame()
```

```python
## Reading Scores by Grade

ninth_reading_s = ninth_df.groupby("school_name")["reading_score"].mean()

tenth_reading_s = tenth_df.groupby("school_name")["reading_score"].mean()

eleventh_reading_s = eleventh_df.groupby("school_name")["reading_score"].mean()

twelveth_reading_s = twelveth_df.groupby("school_name")["reading_score"].mean()

reading_bygrade_df = pd.DataFrame({"9th": ninth_reading_s,
                                "10th": tenth_reading_s,
                                "11th": eleventh_reading_s,
                                "12th": twelveth_reading_s})
reading_bygrade_df
```

```python
## Scores by School Spending

scores_by_spending_df = school_summary_df.loc[:, ["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "Overall Passing Rate"]]
scores_by_spending_df["Per Student Budget Bins"] = pd.cut(x=school_summary_df["Per Student Budget"],
                                                          bins=[0, 585, 615, 645, 675], 
                                                          labels=["<$585", "$585-615", "$615-645", "$645-675"]).to_frame()
scores_by_spending_df = scores_by_spending_df.groupby("Per Student Budget Bins").mean()
scores_by_spending_df
```

```python
## Scores by School Size

scores_by_size_df = school_summary_df.loc[:, ["Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "Overall Passing Rate"]]
scores_by_size_df["School Size Bins"] = pd.cut(x=school_summary_df["Size"],
                                                          bins=[0, 1000, 2000, 5000], 
                                                          labels=["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]).to_frame()
scores_by_size_df = scores_by_size_df.groupby("School Size Bins").mean()
scores_by_size_df
```

```python
## Scores by School Type

scores_by_type_df = school_summary_df.loc[:, ["Type", "Average Math Score", "Average Reading Score", "% Passing Math", "% Passing Reading", "Overall Passing Rate"]]

scores_by_type_df = scores_by_type_df.groupby("Type").mean()

scores_by_type_df
```

```python
## Observable Trends
# 1. Overall passing rate is higher for schools that spend less per student.
# 2. Overall passing rate is higher for small and medium size schools than large schools.
# 3. Overall passing rate is higher for charter schools than district schools.
```
